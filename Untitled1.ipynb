```python
#1. Area of a Circle
import math

def circle_area(r):
    return math.pi * r * r

radius = 5
print("Area of circle:", circle_area(radius))
```

```python
#2. Largest of Three Numbers
def largest(a, b, c):
    return max(a, b, c)

print("Largest number:", largest(10, 25, 15))
```

```python
#3. String Length (without len)
def string_length(s):
    count = 0
    for char in s:
        count += 1
    return count

text = "Hello"
print("Length:", string_length(text))
```

```python
#4. Prime Number Check
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = 11
print(num, "is prime:", is_prime(num))
```

```python
#5. Recursive Fibonacci Number
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("Fibonacci(6):", fibonacci(6))
```

```python
#6. Reverse String Recursively
def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

text = "Python"
print("Reversed:", reverse_string(text))
```

```python
#7. Sum from 1 to n (for loop)
n = 5
total = 0

for i in range(1, n+1):
    total += i

print("Sum:", total)
```

```python
#8. Count Even Numbers in a List
numbers = [1, 2, 3, 4, 5, 6, 8]
count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even numbers count:", count)
```

```python
#9. Prime Numbers from 1 to 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```
